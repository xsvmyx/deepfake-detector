### Checking image dimensions in the dataset:

```python
import os
from PIL import Image

def get_image_dimensions(image_folder):
    dimensions = set()
    max_width = 0
    max_height = 0

    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(image_folder, filename)
            try:
                with Image.open(image_path) as img:
                    w, h = img.size
                    dimensions.add((w, h))
                    if w > max_width:
                        max_width = w
                    if h > max_height:
                        max_height = h
            except:
                print(f"Erreur avec {image_path}")

    return dimensions, (max_width, max_height)

folders = [
    '../images/train/fake',
    '../images/train/real',
    '../images/test/fake',
    '../images/test/real'
]

for folder in folders:
    dims, max_dim = get_image_dimensions(folder)
    print(f"--- {folder} ---")
    print(f"Unique image dimensions: {dims}")
    print(f"Max dimensions: {max_dim}\n")
```

### count images with width and height <= 256px 

```python
import os
from PIL import Image

Image.MAX_IMAGE_PIXELS = None 

DATA_ROOT = "../images"
TARGET_MIN = 256

def check_dataset():
    paths = {
        "Train Fake": os.path.join(DATA_ROOT, 'train/fake'),
        "Train Real": os.path.join(DATA_ROOT, 'train/real'),
        "Test Fake": os.path.join(DATA_ROOT, 'test/fake'),
        "Test Real": os.path.join(DATA_ROOT, 'test/real')
    }

    print(f" Analyse des images (Seuil min: {TARGET_MIN}x{TARGET_MIN}px)...")
    
    global_stats = {}

    for label, folder in paths.items():
        count_valid = 0
        count_total = 0
        
        if not os.path.exists(folder):
            print(f" Dossier non trouv : {folder}")
            continue

        files = [f for f in os.listdir(folder) if f.lower().endswith(('png', 'jpg', 'jpeg', 'webp'))]
        
        for filename in files:
            count_total += 1
            file_path = os.path.join(folder, filename)
            try:
                with Image.open(file_path) as img:
                    w, h = img.size
                    if w >= TARGET_MIN and h >= TARGET_MIN:
                        count_valid += 1
            except:
                continue # Image corrompue ou illisible
        
        global_stats[label] = (count_valid, count_total)
        print(f" {label} : {count_valid} valides sur {count_total}")

    print("\n--- BILAN FINAL ---")
    total_v = sum(v[0] for v in global_stats.values())
    print(f"Total d'images exploitables pour l'entranement : {total_v}")

check_dataset()
```

### performing center crop + resize en place on big images

```python
import os
from PIL import Image
from tqdm import tqdm


Image.MAX_IMAGE_PIXELS = None 

DATA_ROOT = "../images"
TARGET_SIZE = (256, 256)

def preprocess_on_place():
    folders = [
        os.path.join(DATA_ROOT, 'train/fake'),
        os.path.join(DATA_ROOT, 'train/real'),
        os.path.join(DATA_ROOT, 'test/fake'),
        os.path.join(DATA_ROOT, 'test/real')
    ]

    for folder in folders:
        if not os.path.exists(folder):
            continue
            
        print(f"\n Traitement de : {folder}")
        files = [f for f in os.listdir(folder) if f.lower().endswith(('png', 'jpg', 'jpeg', 'webp'))]
        
        for filename in tqdm(files):
            file_path = os.path.join(folder, filename)
            try:
                with Image.open(file_path) as img:
                    # 1. Conversion en RGB (essentiel pour les JPG et le CNN)
                    if img.mode != 'RGB':
                        img = img.convert('RGB')
                    
                    # 2. Calcul du Center Crop
                    w, h = img.size
                    min_dim = min(w, h)
                    left = (w - min_dim) / 2
                    top = (h - min_dim) / 2
                    right = (w + min_dim) / 2
                    bottom = (h + min_dim) / 2
                    
                    img = img.crop((left, top, right, bottom))
                    
                    # 3. Redimensionnement final
                    img = img.resize(TARGET_SIZE, Image.Resampling.LANCZOS)
                    
                    # 4. Sauvegarde (on remplace l'original par un JPG propre)
                    # On change l'extension en .jpg si ncessaire pour l'uniformit
                    base = os.path.splitext(file_path)[0]
                    new_path = base + ".jpg"
                    
                    img.save(new_path, "JPEG", quality=90)
                    
                    # Si l'ancien fichier avait une extension diffrente (ex: .png), on le supprime
                    if file_path != new_path:
                        os.remove(file_path)
                        
            except Exception as e:
                print(f" Erreur sur {filename}: {e}")

if __name__ == "__main__":
    confirm = input(" Ce script va modifier DEFINITIVEMENT tes images en 256x256. Tape 'OK' pour continuer : ")
    if confirm.upper() == "OK":
        preprocess_on_place()
        print("\n Preprocessing termin ! Ton dataset est prt pour le CNN.")
```

### delete the corrupted files based on the logs

```python
import os



corrupted_files = {
    'train/fake': ['8022.jpg', '12854.jpg'],
    'train/real': ['21610.jpg', '15963.jpg', '16011.jpg', '0038.jpg', '13021.jpg', '20964.jpg', '12094.jpg'],
    'test/real': ['5325.jpg', '5197.jpg', '5879.jpg']
}

DATA_ROOT = "../images"

def purge_blacklist(blacklist):
    print(" Purge des fichiers corrompus en cours...")
    count = 0
    
    for folder, files in blacklist.items():
        for filename in files:
            file_path = os.path.join(DATA_ROOT, folder, filename)
            
            if os.path.exists(file_path):
                try:
                    os.remove(file_path)
                    print(f" Supprim : {folder}/{filename}")
                    count += 1
                except Exception as e:
                    print(f" Impossible de supprimer {filename}: {e}")
            else:
                # Peut-tre dj supprim ou extension change
                print(f" Dj absent : {folder}/{filename}")

    print(f"\n Termin. {count} fichiers ont t retirs du dataset.")

purge_blacklist(corrupted_files)
```

### Splitting the Test folder into Val and Test Folders since it was quiet large

```python
import os
import shutil
import random

# Chemins
BASE_DIR = "../images"
TEST_DIR = os.path.join(BASE_DIR, "test")
VAL_DIR = os.path.join(BASE_DIR, "val")

def split_test_to_val():
    # 1. Crer la structure du dossier val
    classes = ['fake', 'real']
    for cls in classes:
        os.makedirs(os.path.join(VAL_DIR, cls), exist_ok=True)

    print(" Dbut du split Test -> Val (50/50)...")

    for cls in classes:
        source_path = os.path.join(TEST_DIR, cls)
        target_path = os.path.join(VAL_DIR, cls)
        
        # Liste tous les fichiers du dossier test/classe
        files = os.listdir(source_path)
        # On mlange pour ne pas prendre que les premiers noms
        random.shuffle(files)
        
        # On calcule 50%
        split_idx = len(files) // 2
        files_to_move = files[:split_idx]
        
        print(f" Dplacement de {len(files_to_move)} images depuis {cls}...")
        
        for f in files_to_move:
            shutil.move(os.path.join(source_path, f), os.path.join(target_path, f))

    print("\n Termin ! Ton dataset est maintenant divis :")
    print(f" {VAL_DIR} contient maintenant la moiti des images de test.")

if __name__ == "__main__":
    split_test_to_val()
```
